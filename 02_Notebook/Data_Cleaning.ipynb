# Cleaning the 'Salary____Estimate' column

```python
# salary column:
#important
#37k - 	66K (Glassdoor est.) ----remove (Glassdoor est.)
#remove $
```

```python
df['Salary____Estimate']
```

```python
# removing the character part of the column
df['Salary____Estimate'] = df['Salary____Estimate'].str.replace(r'\(.*?\)', '', regex=True)
```

```python
# replace / remove the dollar symbol
df['Salary____Estimate'] = df['Salary____Estimate'].str.replace('$','')
```

```python
# filling /replacing the 'K' with 000
df['Salary____Estimate'] = df['Salary____Estimate'].str.replace('K', '000')
```

```python
#verify
df['Salary____Estimate']
```

```python
#-- salary column
#....37000-66000 getting a acutal salary number out of this range
#...coversion of it from object to numeric
#....normalise the value to avg
```

```python
df[['min','max']] = df['Salary____Estimate'].str.split('-', expand=True)
```

```python
df.head(1)
```

```python
df.info()
```

```python
# change min max columns into int for getting salary calculations
```

```python
df['min'] = pd.to_numeric(df['min'])
```

```python
df['max'] = pd.to_numeric(df['max'])
```

```python
#
```

```python
df['salary_cal'] = (df['min'] + df['max']) / 2
```

```python
df['salary_cal']
```

# remove the min max columns

```python
df= df.drop(columns = ['min','max'])
```
